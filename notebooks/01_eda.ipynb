# Beauty Retail Intelligence

Exploratory Data Analysis (EDA)

This notebook explores the Sephora products dataset to understand its structure, evaluate data quality and identify relevant variables before performing business analysis.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

## 2. Load Dataset

```python
products = pd.read_csv("../data/product_info.csv")
```

## 3. Initial Exploration

```python
products.head()
```

### Dataset Information

```python
products.info()
```

### Descriptive Statistics

```python
products.describe()
```

### Missing Values

```python
products.isnull().sum()
```

### Duplicate Records

Check whether the dataset contains duplicated products.

```python
products.duplicated().sum()
```

**Insights**

- No duplicate records were found.
- Each row represents a unique product.
- No duplicate removal is required.

### Product Identifier Validation

Verify whether every product has a unique identifier.

```python
products["product_id"].nunique()
```

**Insights**

- The dataset contains 8,494 unique product IDs.
- Every product has its own identifier, making it suitable for aggregation and relational analysis.

### Brand Coverage

Explore the diversity of brands included in the dataset.

```python
products["brand_name"].nunique()
```

**Insights**

- The dataset includes 304 different brands.
- This provides enough diversity for meaningful brand-level comparisons.

### Product Categories

Analyze the distribution of products across the main categories.

```python
products["primary_category"].value_counts()
```

**Insights**

- Skincare and Makeup dominate the catalog, representing most of the available products.
- Categories such as Gifts and Men contain significantly fewer products.
- This imbalance should be considered when comparing category performance.

### Largest Brands by Number of Products

Identify the brands with the largest product portfolios.

```python
products["brand_name"].value_counts().head(10)
```

**Insights**

- SEPHORA COLLECTION has the largest catalog in the dataset.
- Several premium brands also maintain broad product portfolios.
- A larger catalog does not necessarily imply better customer engagement, which will be explored later.

### Price Distribution

Understand the overall pricing structure of the catalog.

```python
products["price_usd"].describe()
```

**Insights**

- Most products are priced between $25 and $58.
- The average price is approximately $52.
- Products priced at $1,900 indicate the presence of extreme outliers that should be investigated before visualization.

### Products Without Ratings

Identify products that have not yet received customer ratings.

```python
products[products["rating"].isna()].shape
```

**Insights**

- Only 278 products have no rating.
- Missing ratings likely indicate insufficient customer feedback rather than data quality issues.
- These products will be excluded only from analyses that require customer ratings.

### Invalid Prices

Check whether any products have invalid prices.

```python
products[products["price_usd"] <= 0]
```

### Rating Validation

Ensure all product ratings fall within the expected range.

```python
products[(products["rating"] < 1) | (products["rating"] > 5)]
```

### Review Count Validation

Check for invalid review counts.

```python
products[products["reviews"] < 0]
```

### Engagement Validation

Verify that engagement metrics contain valid values.

```python
products[products["loves_count"] < 0]
```

### Category Validation

Identify products with missing category information.

```python
products[products["primary_category"].isna()]
```

## Data Quality Summary

The dataset presents a high level of data quality. No duplicated records or invalid numerical values were identified. Missing values are primarily associated with optional product attributes (e.g., sale price, variations, ingredients) rather than data quality issues.

Therefore, only minimal preprocessing is required before conducting the business analysis.
